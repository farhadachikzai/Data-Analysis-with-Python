```python
import pandas as pd
import numpy as np
import sqlite3 
import matplotlib.pyplot as plt
```

```python
conn = sqlite3.connect('archive/sqlite-sakila.db')

df = pd.read_sql('''
    SELECT
        rental.rental_id, rental.rental_date, rental.return_date,
        customer.last_name AS customer_lastname,
        store.store_id,
        city.city AS rental_store_city,
        film.title AS film_title, film.rental_duration AS film_rental_duration,
        film.rental_rate AS film_rental_rate, film.replacement_cost AS film_replacement_cost,
        film.rating AS film_rating
    FROM rental
    INNER JOIN customer ON rental.customer_id == customer.customer_id
    INNER JOIN inventory ON rental.inventory_id == inventory.inventory_id
    INNER JOIN store ON inventory.store_id == store.store_id
    INNER JOIN address ON store.address_id == address.address_id
    INNER JOIN city ON address.city_id == city.city_id
    INNER JOIN film ON inventory.film_id == film.film_id
    ;
''', conn, index_col='rental_id', parse_dates=['rental_date', 'return_date'])
```

What is the common Film Duration 

```python
colors = ["#005073", "#107dac", "#189ad3", "#1ebbd7", "#71c7ec"]
ax = (df["film_rental_duration"]
      .value_counts()
      .plot(kind="bar", 
            figsize=(14,6),
            color=colors
            )
        )

for container in ax.containers:
    ax.bar_label(container)

ax.set_xlabel("Film Duration")
ax.set_ylabel("Number of Films")
ax.set_title("Common Duration of Rental Films")
```

calculate a new rental_days

```python
df["rental_date"] = pd.to_datetime(df["rental_date"])
df["return_date"] = pd.to_datetime(df["return_date"])
```

```python
df["rental_days"] = (
    df["return_date"] - df["rental_date"]
).dt.days
```

```python
ax = df["rental_days"].plot(kind="density")

ax.axvline(df["rental_days"].mean(), color="red")
```

```python
df["film_daily_rental_rate"] = df["film_rental_rate"] / df["film_rental_duration"]
```

Films daily rental rate density plot

```python
ax = df["film_daily_rental_rate"].plot(kind="density", figsize=(10,6))

ax.axvline(df["film_daily_rental_rate"].mean(), color="red")
```
