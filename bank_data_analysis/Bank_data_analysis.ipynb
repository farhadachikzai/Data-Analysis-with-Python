```python
# the y column actually mean if a customer converted to easily manipolate data we need to change yes and no to 0 and 1

bank_data['converted'] = bank_data['y'].apply(lambda x: 0 if x=='no' else 1)
```

```python
# converted and balance baxplot with outliers  
ax = bank_data[['converted', 'balance']].boxplot(by='converted', showfliers=True, figsize=(7,5))

ax.set_ylabel('Average of bank blance')
ax.set_xlabel('Conversion')
ax.set_title('Average Bank Balance and Conversion')

plt.suptitle("")
plt.show()
```

```python
#converted and balance without outliers 
ax = bank_data[['converted', 'balance']].boxplot(by='converted', showfliers=False, figsize=(9,5))
ax.set_xlabel('Converted')
ax.set_ylabel('Average Bank Balance')
ax.set_title('Average Bank Balance by converstion (0 = no conversion and 1 = conversion)')

plt.suptitle("")
plt.show()
```

```python
#lets visualize converted and bank balance using violineplot

fig, axes = plt.subplots(figsize=(10,6))
sns.violinplot('converted', 'balance', data=bank_data, axis=axes)

axes.set_xlabel('Converted')
axes.set_ylabel('Average Balance')
axes.set_title('Bank Balance and Conversion')
```

```python
percentage_campaign = bank_data.groupby('campaign')['converted'].sum() / bank_data.groupby('campaign')['converted'].count()*100
#top 10 campaign which had pasitive result onn converting customer
percentage_campaign.head(10)
```

```python
#conversion plot by campaign

percentage_campaign.plot(kind='bar', figsize=(15,6), color='coral', grid=True)

plt.xlabel('Campaign')
plt.ylabel('Percentage % of Conversion')
plt.title('Percentage of conversion over Number of Campaign contacts', fontweight='bold')

plt.show()
```

```python
#number of customer converted compare to thier job
conversion_by_job = bank_data.groupby('job')['converted'].sum().plot(kind='bar', figsize=(12,6), color='coral')

plt.xlabel('Job Category')
plt.ylabel('Number of converted Customer')
plt.title('Number of converted Custober by Job')

plt.xticks(rotation=30)
```

```python
# Percentage of conversion compare to customer designation
conversion_by_job_per = round(bank_data.groupby('job')['converted'].sum() / bank_data.groupby('job')['converted'].count()*100, 2)
```

```python
#job conversion percentage 
conversion_by_job_per.plot(kind='bar', figsize=(12, 6), color='blue')
plt.xlabel('Job Category')
plt.ylabel('Conversion Percentage %')
plt.title('Percentage % of conversion over Job Category', fontweight='bold', fontsize=14)
plt.xticks(rotation=30, fontweight='bold')
```
