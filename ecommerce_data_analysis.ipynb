```python
#removing missing values
df_no_missing = df.dropna()
```

```python
df_no_missing.isnull().sum().sort_values(ascending=False)
```

```python
df_no_missing['CustomerID'] = df_no_missing['CustomerID'].astype('int64')
```

```python
df_no_missing.info()
```

```python
#copying the data set to make short 
df2 = df_no_missing.copy()
```

```python
df2.describe().round(2)
```

```python
#voilen plot for quantity
sns.set(style='whitegrid')
ax = sns.violinplot(df2['Quantity'])
```

```python
#voilin plot greater than 0
df3 = df2[df2.Quantity > 0]
```

```python
df3
```

```python
# violinplot to understan the quantity of items
sns.set(style='whitegrid')
ax = sns.violinplot(df3['Quantity'])
```

```python
#to understand the total amount by creating new column as quantity * uniteprice

df2['amoun_spent'] = df2['Quantity'] * df2['UnitPrice']
```

```python
df2
```

```python
#for the yearly sales we need to create year month column

df2['year_month'] = df2['InvoiceDate'].dt.to_period('M')
```

```python
#sales by customber 
sales_per_cust = df2.groupby(['CustomerID', 'Country'], as_index=False)['InvoiceNo'].count().sort_values('InvoiceNo', ascending=False)
```

```python
sales_per_cust.head(10)
```

```python
sales_mount_country = df2.groupby(['CustomerID', 'Country'], as_index=False)['amoun_spent'].sum().sort_values('amoun_spent', ascending=False)
```

```python
top_10_country = sales_mount_country.head(15)
top_10_country
```

```python
#most sales by country
plt.bar(x=top_10_country['Country'], height=top_10_country['amoun_spent'])
```

```python
#customer Orders
orders = df2.groupby(['CustomerID'], as_index=False)['InvoiceNo'].count()

plt.subplots(figsize=(15,5))
plt.plot(orders.CustomerID, orders.InvoiceNo)

plt.xlabel('Customer ')
plt.ylabel('Number of Sales')

plt.title('Best Customer Sales', fontweight='bold')
```

```python
#which customer spent more money

money_spent = df2.groupby(['CustomerID', 'Country'], as_index=False)['amoun_spent'].sum()
money_spent
```

```python
#which customer spent more money
plt.subplots(figsize=(15,5))
plt.plot(money_spent.CustomerID, money_spent.amoun_spent)
```

```python
#creating 2 columns year_month and hours but data will get from invoiceDate

df2.insert(loc=2, column='yearmonth', value=df2['InvoiceDate'].map(lambda x: 100*x.year + x.month))
```

```python
df2.insert(loc=5, column='hour', value=df2['InvoiceDate'].dt.hour)
```

```python
df2
```

```python
#number of orders per month 
df2.groupby('InvoiceNo')['yearmonth'].unique().value_counts().sort_index().plot(kind='bar', figsize=(15,6))

plt.xlabel('Year Month')
plt.ylabel('Number of Sales')
plt.title('Number of items sold Monthly', fontweight='bold')
plt.xticks(rotation=45)
```

```python
df2.insert(loc=6, column='weekday', value=df2['InvoiceDate'].dt.weekday)
```

```python
#number of sales by week days
ax = df2.groupby('InvoiceNo')['weekday'].unique().value_counts().sort_index().plot(kind='bar', figsize=(15,5))
plt.xlabel('Days')
plt.ylabel('Number of Orders')
plt.title('Daily Sales Order', fontweight='bold')
ax.set_xticklabels(('Mon', 'Tues', 'Wed', 'Thur', 'Fri', 'Sund'), rotation=45)

plt.show()
```

```python
#hourly Sales 
ax = df2.groupby('InvoiceNo')['hour'].unique().value_counts().iloc[:-1].sort_index().plot(kind='bar', figsize=(15,5))

plt.xlabel('Hours of the Day')
plt.ylabel('Number of Invoice')
plt.title('Hourly Order ', fontweight='bold')

ax.set_xticklabels(('6:00', '7:00', '8:00','9:00','10:00','11:00','12:00','13:00','14:00','15:00','16:00','17:00','18:00','19:00','20:00'), 
rotation=45)
```

```python
df2.insert(loc=5, column='weekofyear', value=df2['InvoiceDate'].dt.weekofyear)
```

```python
df2.groupby('InvoiceNo')['weekofyear'].unique().value_counts().iloc[:-1].sort_index().plot(kind='bar', figsize=(18,6))
plt.xlabel('Week of the Year')
plt.ylabel('Number of Order')
plt.title('Weekly Order During 2011', fontweight='bold', fontsize=14)
plt.xticks(rotation=45)
```

```python
#analysing Unit Price
df2.UnitPrice.describe()
```

```python
#understanding items price
plt.subplots(figsize=(15, 6))
sns.boxplot(df2.UnitPrice)
plt.show()
```

```python
#checking if we have item without cost
item_free = df2[df2['UnitPrice'] == 0]
item_free.head()
```

```python
item_free.groupby('yearmonth')['yearmonth'].count()
```

```python
#ploting to know in which month we have give more free items
ax = item_free.groupby('yearmonth')['yearmonth'].count().sort_index().plot(kind='bar', figsize=(15,6))

plt.title('Number of Free Items was given over the Year to the Customer', fontweight='bold')
plt.xlabel('Month')
plt.ylabel('Number of Free items')

ax.set_xticklabels(('Dec-10', 'Jan-11','Feb-11', 'Mar-11', 'Apr-11', 'May-11', 'Jul-11', 'Aug-11', 
                    'Sep-11', 'Oct-11', 'Nov-11'), rotation=30)
```

```python
#analysing data per Country 
ax = df2.groupby('Country')['InvoiceNo'].count().sort_values().plot(kind='barh', figsize=(15, 8))
plt.title('Number of Orders per Country', fontweight='bold', fontsize=14)
plt.xlabel('Number of Orders')
plt.ylabel('Countries')

plt.show()
```

```python
#in order to fully visualize other countries we have to remove the United kindom 

order_by_country = df2.groupby('Country')['InvoiceNo'].count().sort_values()
del order_by_country['United Kingdom']
#now the united kingdom was deleted we can plot 
order_by_country.plot(kind='barh', figsize=(15, 10))
plt.title('Number of Order per Country except United kindom', fontsize=14, fontweight='bold')
plt.xlabel('Number of Items')
plt.ylabel('Countries')
plt.show()
```

```python
#amount expent by each country excluding UK
amount_expent_country = df2.groupby('Country')['amoun_spent'].sum().sort_values()
del amount_expent_country['United Kingdom']
amount_expent_country.plot(kind='barh', figsize=(15, 10))
plt.title('Amount spent by Each country', fontsize=14, fontweight='bold')
plt.xlabel('Amount in $')
plt.ylabel('Countries')
plt.show()
```
